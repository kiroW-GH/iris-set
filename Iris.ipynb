```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from torch.utils.data import Dataset
```

Loading the data

```python
iris_csv = pd.read_csv('data\Iris\Iris.csv')
```

```python
iris_csv[50:100]
```

```python
x =iris_csv[['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm',	'PetalWidthCm']]#get rid of id as it is just wasted compute
```

```python
y = iris_csv['Species']
for index, i in enumerate(y):
    if i == 'Iris-setosa':
        y[index] = [1,0,0]
    if i == 'Iris-virginica':
        y[index] = [0,0,1]
    elif i == 'Iris-versicolor':
        y[index] = [0,1,0]
```

```python
y
```

```python
import numpy as np
# let's turn the labels into classes
```

```python
y = torch.tensor(y)
x= torch.tensor(np.array(x))
```

```python
x.dtype
x =x.to(torch.float32)
```

```python
model = nn.Sequential(nn.Linear(4, 10),
                      nn.ReLU(),
                      nn.Linear(10,3))
print(x.dtype, x.shape)
model(x)
```

```python
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
x.shape
y.shape #so to have the same inner shapes
```

```python
y=y.argmax(dim=1) #gets the indice of the correct label
```

```python
train_data = DataLoader(TensorDataset(x,y), shuffle=True, batch_size=5)#put our dataset with specifying what are the inputs and outputs
```

```python
next(iter(train_data))
```

```python
import matplotlib.pyplot as plt
```

```python
for idx, idy in train_data:
    output = model(idx)
    loss = loss_fn(output, idy)#loss takes the index of the class only
    print('loss is ', loss)
    optim.zero_grad()
    loss.backward()
    optim.step()
```
